### Create frequent datasets

```python
from  mlxtend.preprocessing import TransactionEncoder
```

```python
t=TransactionEncoder()
```

```python
t_arr=t.fit_transform(data)
```

```python
t_arr
```

### create the frequent itemsets


```python
from mlxtend.frequent_patterns import apriori
```

```python
freq_items=apriori(data,min_support=0.1,use_colnames=True)
```

```python
freq_items
```

### from frequent itemsets extract the association rules


```python
from mlxtend.frequent_patterns import association_rules
```

```python
assoc_rules=association_rules(freq_items,min_threshold=1.0)
assoc_rules.head()
```

```python
assoc_rules[assoc_rules['lift'] > 1]
```

```python
plt.scatter(assoc_rules['support'],assoc_rules['confidence'])
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.show()
```

```python
assoc_rules[assoc_rules['lift'] > 2]
```

```python
assoc_rules1=association_rules(freq_items,metric = 'lift',min_threshold=0.9)
```

```python
movie_count=[]
col=data1.columns
for columns in col:
    count = (data1[columns] == 1).sum()
    movie_count.append(count)
```

```python
plt.figure(figsize=(10, 10))
plt.bar(col, movie_count)
plt.xlabel('Column Names')
plt.ylabel('Count of 1')
plt.title('Count of 1 in Each Column')
plt.xticks(rotation=60)  
plt.tight_layout()  
plt.show()
```

```python
freq_movies=apriori(data1,min_support=0.1,use_colnames=True)
freq_movies.head()
```

```python
assoc_rules1 = association_rules(freq_movies, metric="lift", min_threshold=1)
assoc_rules1.head()
```
